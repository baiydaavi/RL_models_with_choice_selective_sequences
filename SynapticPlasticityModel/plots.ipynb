# Download recorded data
Download sequential and synchronous data for PL-NAc and sequential data for Thal-NAc if recorded data folder does not already exist.

```python
if not os.path.exists('recorded_data'):
    url = 'https://zenodo.org/record/5784868/files/recorded_data.zip'
    r = requests.get(url, allow_redirects=True)
    open('recorded_data.zip', 'wb').write(r.content)

    with zipfile.ZipFile("recorded_data.zip","r") as zip_ref:
        zip_ref.extractall(".")

    os.remove("recorded_data.zip")
```
