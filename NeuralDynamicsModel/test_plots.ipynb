# Download testing data
Download and unzip the test data if the testing data folder does not exist already.

```python
if not os.path.exists('testing_data'):
    url = 'https://zenodo.org/record/5774370/files/testing_data.zip'
    r = requests.get(url, allow_redirects=True)
    open('testing_data.zip', 'wb').write(r.content)

    with zipfile.ZipFile("testing_data.zip","r") as zip_ref:
        zip_ref.extractall("testing_data")
```
